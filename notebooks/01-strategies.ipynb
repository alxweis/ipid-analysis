# IPID selection-strategy distribution

Loads the classified strategies (`strategies.pq`) for one measurement, computes
the percentage share of each IPID selection strategy, plots it as a bar chart
and saves the figure to `reports/figures/` as a PDF.

The heavy lifting lives in the package (`ipid_analysis.plots`); this notebook
just wires it together.

```python
%load_ext autoreload
%autoreload 2

from pathlib import Path

from ipid_analysis.config import FIGURES_DIR, PROCESSED_DATA_DIR
from ipid_analysis.plots import plot_strategy_distribution, strategy_distribution
```

## Select the measurement

Use the same key you passed to `classify.py`.

```python
measurement = "ipid/icmp_2026-06-29_15-56-56"

strategies_path = PROCESSED_DATA_DIR / measurement / "strategies.pq"
output_path = FIGURES_DIR / f"{Path(measurement).name}_strategy_distribution.pdf"
```

## Distribution

`strategy_distribution` runs a streaming `GROUP BY` in DuckDB, so it works even
on the full (>100 GB) output without loading it into memory.

```python
dist = strategy_distribution(strategies_path)
dist
```

## Plot and save as PDF

```python
fig = plot_strategy_distribution(
    dist, output_path, title=f"IPID strategy distribution — {Path(measurement).name}"
)
print(f"saved -> {output_path}")
```

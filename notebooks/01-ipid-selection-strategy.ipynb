# IPID Selection-Strategy Distribution

## Input

```python
measurement = "ipid/tcp-80_YYYY-MM-DD_HH-MM-SS"
```

## Setup

```python
%load_ext autoreload
%autoreload 2

from pathlib import Path

import typer
from loguru import logger

from ipid_analysis.config import IPID_STRATEGY_DATA_NAME, PROCESSED_DATA_DIR
from ipid_analysis.plots import strategy_counts, strategy_percentages, info_strategy_distribution, \
    plot_strategy_distribution

strategies_path = PROCESSED_DATA_DIR / Path(measurement) / IPID_STRATEGY_DATA_NAME
if not strategies_path.is_file():
    logger.error(f"not found: {strategies_path}")
    raise typer.Exit(code=1)

counts = strategy_counts(strategies_path)
total = sum(counts.values())
percentages = strategy_percentages(counts, total)
```

## Distribution

```python
info = info_strategy_distribution(strategies_path, counts, total, percentages)
```

## Plot

```python
plot, plot_path = plot_strategy_distribution(strategies_path, percentages)
print(f"figure saved -> {plot_path}")
```
